```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import interpolate
from scipy import stats
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
e = [0.5, 1., 1.5, 2.]
plt.errorbar(x, y, yerr=e, fmt='s')
plt.show()
```

```python
index = np.arange(5)
values = [5, 6, 3, 4, 6]
SD = [0.8, 2, 0.4, 0.9, 1.3]
plt.title('A Bar Chart')
plt.bar(index, values, yerr = SD, error_kw = {'ecolor' : '0.8', 'capsize' :14}, alpha=0.7, label = 'First')
plt.xticks(index+0.2,['a', 'b', 'c', 'd', 'e'])
plt.legend(loc=2)
plt.show()
```
